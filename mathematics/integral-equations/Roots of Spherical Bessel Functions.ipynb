```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.special as sp
```

```python
k = (1+0.0j)
# k = 1
rng = np.linspace(0, 10*np.pi, 1000)
n = 1
jn = sp.spherical_jn(n, k*rng)
# zeros = sp.jn_zeros(n, 10)
```

```python
fig, ax = plt.subplots()

plt.rcParams['text.usetex'] = False
x = np.array([x.real for x in jn])
y = np.array([x.imag for x in jn])
ax.plot(rng, x, label='real')
ax.plot(rng, y, label='imag')
# ax.scatter(zeros, np.zeros(len(zeros)))
ax.legend()
ax.set_title('$j_1(kr)$ for $r \in [0, 10\pi]$', )
# ax.scatter(x, y)
ax.set_xlabel('r')
ax.set_ylabel('Magnitude')
plt.show()
```
