```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.special as sp
```

```python
k = (2+1j)
# k = 1
rng = np.linspace(0, 10*np.pi, 1000)
n = 1
jn = sp.spherical_jn(n, k*rng)
```

```python
fig, ax = plt.subplots()

x = np.array([x.real for x in jn])
y = np.array([x.imag for x in jn])
ax.plot(x)
ax.plot(y)
# ax.scatter(x, y)
```

```python
fig, ax = plt.subplots()
mag = np.sqrt(x**2+y**2)
ax.plot(mag)
ax.set_yscale('log')
```
